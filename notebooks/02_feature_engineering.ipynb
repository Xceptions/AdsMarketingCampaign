```python
import numpy as np
import pandas as pd
```

```python
df_train = pd.read_csv('/Users/macbookair/Documents/GitHub/AdsMarketingCampaign/data/processed/ads_campaign_data_train.csv')
df_test = pd.read_csv('/Users/macbookair/Documents/GitHub/AdsMarketingCampaign/data/processed/ads_campaign_data_test.csv')
df_train.head(4)
```

### Feature Engineering

```python
df_train['Clicks_Plus_Leads'] = df_train['Clicks'] + df_train['Leads']
df_test['Clicks_Plus_Leads'] = df_test['Clicks'] + df_test['Leads']
```

### Feature Selection

```python
features = ['Clicks', 'Leads', 'Clicks_Plus_Leads']
target = ['Revenue']
selection = features + target

df_train = df_train[selection]
df_test = df_test[selection]

df_train.to_csv('/Users/macbookair/Documents/GitHub/AdsMarketingCampaign/data/feature_store/ads_campaign_features_train.csv', index=False)
df_test.to_csv('/Users/macbookair/Documents/GitHub/AdsMarketingCampaign/data/feature_store/ads_campaign_features_test.csv', index=False)
```
